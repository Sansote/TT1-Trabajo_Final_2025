```python
import pandas as pd
import numpy as np
from arch import arch_model
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Importo el repo
!git clone https://github.com/Sansote/TT1-Trabajo_Final_2025.git
```

```python
# Importo el set de datos
ruta = 'TT1-Trabajo_Final_2025/salida_acciones_dolarizadas.xlsx'
df = pd.read_excel(ruta, index_col='Date', parse_dates=True)
df.head()
```

```python
# Me quedo con los retornos
retornos_usd = [col for col in df.columns if "usd_ret_usd" in col or "Date" in col]
retornos_usd
```

```python
retornos = df[retornos_usd]
retornos.head()
```

```python
# Calculo el modelo

resultados_garch = {}
errores = []

for col in retornos:
    try:
        print(f"Procesando columna: {col}")

        serie = retornos[col]
        serie = pd.to_numeric(serie, errors='coerce').dropna().astype(float)

        if serie.empty:
            raise ValueError("Serie vacía después de limpieza")

        if serie.abs().max() > 1e6:
            serie = serie * 1e-15


        modelo = arch_model(serie, vol='Garch', p=1, q=1, mean='zero')
        resultado = modelo.fit(disp='off')


        pred_vol = np.sqrt(resultado.conditional_volatility)

        # Volatilidad real
        vol_real = serie.rolling(window=30).std()


        pred_vol_aligned = pred_vol[29:]
        vol_real_aligned = vol_real[29:]

        # Calculo MAE Y MSE
        mae = mean_absolute_error(vol_real_aligned, pred_vol_aligned)
        mse = mean_squared_error(vol_real_aligned, pred_vol_aligned)

        # Guardar resultados
        resultados_garch[col] = {
            'modelo': resultado,
            'MAE': mae,
            'MSE': mse
        }

    except Exception as e:
        print(f"⚠️ Error en columna '{col}': {e}")
        errores.append((col, str(e)))
```

```python
# Check errores
print("Columnas con errores:")
for col, error in errores:
    print(f"{col}: {error}")
```

```python
resultados_garch
```

```python
# Armo un filtro para ver los parametros significativos. Uso un umbral de 5%
resumen_significancia = []

for col, resultado_dict in resultados_garch.items():
    modelo = resultado_dict['modelo']  # extraer el modelo GARCH
    params = modelo.params
    pvals = modelo.pvalues

    for coef_name in params.index:
        resumen_significancia.append({
            'columna': col,
            'coeficiente': coef_name,
            'valor': params[coef_name],
            'pvalor': pvals[coef_name]
        })

df_significancia = pd.DataFrame(resumen_significancia)
```

```python
df_significativos = df_significancia[df_significancia['pvalor'] < 0.05]
```

```python
# Los guardo
df_significativos.to_excel("df_garch_significativos_1_1.xlsx", index=False)
```

## Trabajo sin outliers

```python
# Saco los outliers definiendolos como los que esta

def eliminar_outliers_iqr(df):
    df_filtrado = df.copy()
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        # Reemplaza los outliers con NaN
        df_filtrado[col] = df[col].where((df[col] >= limite_inferior) & (df[col] <= limite_superior))
    return df_filtrado
```

```python
retornos_sin_ouliers=eliminar_outliers_iqr(retornos)
```

```python
resultados_garch = {}
errores = []

for col in retornos_sin_ouliers:
    try:
        print(f"Procesando columna: {col}")

        serie = retornos_sin_ouliers[col]
        serie = pd.to_numeric(serie, errors='coerce').dropna().astype(float)

        if serie.empty:
            raise ValueError("Serie vacía después de limpieza")

        if serie.abs().max() > 1e6:
            serie = serie * 1e-15


        modelo = arch_model(serie, vol='Garch', p=1, q=1, mean='zero')
        resultado = modelo.fit(disp='off')

        pred_vol = np.sqrt(resultado.conditional_volatility)

        vol_real = serie.rolling(window=30).std()

        pred_vol_aligned = pred_vol[29:]
        vol_real_aligned = vol_real[29:]

        mae = mean_absolute_error(vol_real_aligned, pred_vol_aligned)
        mse = mean_squared_error(vol_real_aligned, pred_vol_aligned)

        resultados_garch[col] = {
            'modelo': resultado,
            'MAE': mae,
            'MSE': mse
        }

    except Exception as e:
        print(f"⚠️ Error en columna '{col}': {e}")
        errores.append((col, str(e)))
```
